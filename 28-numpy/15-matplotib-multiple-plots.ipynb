```python
import matplotlib.pyplot as plt
import numpy as np
```

# Display Multiple Plots
With the subplots( ) function you can draw multiple plots in one figure:

```python
#plot 1:
x = np.array([0 , 1 , 2 , 3 ])
y = np.array([3 , 8 , 1 , 10 ])
plt.subplot(1 , 2 , 1 )
plt.plot(x,y,  marker='P', mec='r',  mfc = 'yellow')

x = np.array([0 , 1 , 2 , 3 ])
y = np.array([10 , 20 , 30 , 40 ])
plt.subplot(1 , 2 , 2 )
plt.plot(x,y,  marker='D', mec='r',  mfc = 'green')
plt.show()
```
