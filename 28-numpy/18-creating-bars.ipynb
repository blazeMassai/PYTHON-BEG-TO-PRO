```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array(["A" , "B" , "C" , "D" ])
y = np.array([3 , 8 , 1 , 10 ])
plt.bar(x,y, color='#8A2BE2')
plt.suptitle('BAR CHART')
plt.show()
```

```python
#horizontal bar
x = np.array(["A" , "B" , "C" , "D" ])
y = np.array([3 , 8 , 1 , 10 ])
plt.barh(x, y, color='#8B0000')
plt.show()
```

```python
#bar width
#The default width value is 0.8
x = np.array(["A" , "B" , "C" , "D" ])
y = np.array([3 , 8 , 1 , 10 ])
plt.bar(x, y, width = 0.4, color='#2E8B57' )
plt.show()
```

```python
#bar height
x = np.array(["A" , "B" , "C" , "D" ])
y = np.array([3 , 8 , 1 , 10 ])
plt.barh(x, y,  height = 0.1)
plt.show()
```
