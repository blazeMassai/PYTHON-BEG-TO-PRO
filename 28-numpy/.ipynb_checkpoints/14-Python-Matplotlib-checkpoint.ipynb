# What is Matplotlib?
Matplotlib is a low level graph plotting library in python that serves as a visualization utility

```python
import matplotlib

print(matplotlib.__version__)
```

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0,6])
x1 = np.array([0,4])
y2 = np.array([0,2])
ypoints = np.array([0, 250])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
a = np.array([1 , 8 ])
b = np.array([3 , 10 ])

plt.plot(a, b)
plt.show()

#plotting without line
plt.plot(a, b, 'o')
plt.show()
```

```python
c = np.array([1 , 2 , 6 , 8 ])
d = np.array([3 , 8 , 1 , 10 ])
plt.plot(c, d)
plt.show()
```

# Default X-Points

If we do not specify the points in the x-axis, they will get the default values 0, 1, 2, 3, (etc. depending on
the length of the y-points.

```python
e = np.array([3 , 8 , 1 , 10 , 5 , 7 ])
plt.plot(e)
plt.show()
```

# Markers
You can use the keyword argument marke r to emphasize each point with a specified marker:

```python
e1 = np.array([3 , 8 , 1 , 10 ])
plt.plot(e1, marker = 'o' )
plt.show()

#Mark each point with a star or plus:
#ms=marker size
plt.plot(e1, marker = 'D',ms=15, color='r' )
plt.show()
```

# Marker Color

You can use the keyword argument marker edge color or the shorter 'mec' to set the color of the edge of
the markers:
You can use the keyword argument marker face color or the shorter mfc to set the color inside the edge
of the markers:

You can use the keyword argument linestyle , or shorter 'ls' , to change the style of the plotted line:
linestyle can be written as ls .
dotted can be written as : 
dashed can be written as - - .
You can also use Hexadecimal color values:

You can use the keyword argument linewidth or the shorter lw to change the width of the line.
The value is a floating number, in points:

```python
f = np.array([3 , 8 , 1 , 10 ])
plt.plot(f, marker = 'o' , ms = 20 , mec = 'r', mfc = 'hotpink' )
plt.show()

plt.plot(f, marker = 'o' , ms = 20 , mec = 'r', mfc = 'green', ls='dotted',  linewidth = '7.5')
plt.show()
```

# Multiple Lines
You can plot as many lines as you like by simply adding more plt.plot( ) functions:

```python
g = np.array([3 , 8 , 1 , 10 ])
j = np.array([6 , 2 , 7 , 11 ])

plt.plot(g)
plt.plot(j)
```

```python
k = np.array([0 , 1 , 2 , 3 ])
l = np.array([3 , 8 , 1 , 10 ])
m = np.array([0 , 1 , 2 , 3 ])
n = np.array([6 , 2 , 7 , 11 ])

plt.plot(k, l, m, n)
plt.show()
```
