SUBPLOTS() FUNCTION

The subplots( ) function takes three arguments that describes the layout of the figure.
The layout is organized in rows and columns, which are represented by the first and second argument.
The third argument represents the index of the current plot.

hii inawezesha kuchora plot mbili kwenye one go usaing same variable names

```python
import matplotlib.pyplot as plt
import numpy as np
# import matplotlib.pyplot as figure
```

```python
# ya juu tunabadiri size
# g = plt.figure()
# g.set_figwidth(10)
# g.set_figheight(10)

x=np.array([0,1,2,3])
y=np.array([3,8,1,10])


plt.subplot(1,2,1)
#the figure has 1 rows, 2 columns, and this plot is the first plot.

plt.plot(x,y, marker='o', ms=10, mfc='r')

# ya chini tunabadiri size
# f = plt.figure()
# f.set_figwidth(10)
# f.set_figheight(10)


x = np.array([0 , 1 , 2 , 4 ])
y = np.array([10 , 20 , 30 , 40 ])
plt.subplot(1, 2 , 2 )
plt.subplots_adjust(wspace=1, hspace=1)
plt.plot(x,y, marker='D', ms=10, mfc='hotpink')

plt.suptitle('MAUZO NA MATUMIZI')
```

```python

a = np.array([0, 1, 2, 3])
b = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 1)
plt.plot(a,b)
plt.title("SALES" )

a = np.array([0, 1, 2, 3])
b = np.array([10, 20, 30, 40])

plt.subplots_adjust(left=0, bottom=0, right=1, top=1, wspace=0.5, hspace=0.5)
# plt.subplots_adjust(left=0, bottom=0, right=1, top=1, wspace=0, hspace=0)

plt.subplot(2, 3, 2)
plt.plot(a,b)
plt.title("ACCOUNTS" )

a = np.array([0, 1, 2, 3])
b = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 3)
plt.plot(a,b)
plt.title("SUPPLIES" )

a = np.array([0, 1, 2, 3])
b = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 4)
plt.plot(a,b)

plt.title("DEBTS" )

a = np.array([0, 1, 2, 3])
b = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 5)
plt.plot(a,b)
plt.title("MISCELENIOUS" )

a = np.array([0, 1, 2, 3])
b = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 6)
plt.plot(a,b)
plt.title("STOCKS" )
# plt.suptitle('MAUZO NA MATUMIZI')

plt.show() 
```
